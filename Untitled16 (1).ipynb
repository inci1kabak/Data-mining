BAZI DEMOGRAFİK BİLGİ İÇEREN SÜTUNLARI İŞE YARAMAYACAĞINI DÜŞÜNDÜĞÜM VE KATEGORİK OLDUKLARI İÇİN KALDIRIYORUM.

```python
#Sütun silme
df.drop(['job', 'marital', 'education', 'contact'], axis=1, inplace=True)
```

```python
df.head()
```

SÜTUNLARIMDA BOŞ DEĞER VAR MI DİYE BAKIYORUM.

```python
# Her sütundaki boş değerleri sayma
boş_değerler = df.isnull().sum()
```

```python
print(boş_değerler)
```

KATEGORİK DEĞERLERİ SAYISAL DEĞERLERE DÖNÜŞTÜRÜYORUM.

```python
# 'default', 'housing', 'y' ve 'loan' sütunlarındaki "no" için 0, "yes" için 1 olarak değiştirme

df['default'] = df['default'].apply(lambda x: 1 if x == 'yes' else 0)
df['housing'] = df['housing'].apply(lambda x: 1 if x == 'yes' else 0)
df['loan'] = df['loan'].apply(lambda x: 1 if x == 'yes' else 0)
# 'y' sütununu 1'e (yes) ve 0'a (no) dönüştürme
df['y'] = df['y'].apply(lambda x: 1 if x == 'yes' else 0)
```

```python
df.head()
```

MONTH DEĞİŞKENİM İÇİN İLK 6 AYI VE SON 6 AYI AYIRARAK SAYISALLAŞTIRMA YAPTIM.

```python
# 'month' sütununu ilk 6 ay ve son 6 ay olmak üzere iki gruba ayırma
df['month'] = df['month'].apply(lambda x: 1 if x in ['jan', 'feb', 'mar', 'apr', 'may', 'jun'] else 2)
```

```python
df.head()
```

POUTCOME SÜTUNUNDA BOŞ DEĞERLER GÖRÜNÜYOR. BU SÜTUNDAKİ BOŞLUKLARI MOD İLE DOLDURMAYA KARAR VERİYORUM. PEKİ BU SÜTUNUN MODU NEDİR?

```python
print(df['poutcome'].value_counts())
```

```python
poutcome_mode = df['poutcome'].mode()[0]
print("poutcome sütununun modu (en çok tekrarlanan değer):", poutcome_mode)
```

```python
# 'poutcome' sütunundaki eksik değerleri en sık tekrarlanan değerle doldurma
df['poutcome'] = df['poutcome'].fillna(df['poutcome'].mode()[0])
```

```python
df.head()
```

```python
# One-hot encoding
poutcome_encoded = pd.get_dummies(df['poutcome'], prefix='poutcome')

# Dönüştürülen sütunları asıl veri çerçevesine eklemek
df = pd.concat([df, poutcome_encoded], axis=1)
```

```python
df.head()
```

POUTCOME SÜTUNUNU SİLMEK GEREKİYOR.

```python
df.drop('poutcome', axis=1, inplace=True)
```

```python
df.head()
```

LOJİSTİK REGRESYON

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report



# Bağımsız değişkenler (X) ve bağımlı değişken (y) ayırma
X = df.drop(columns=['y'])  # Bağımsız değişkenler
y = df['y']  # Bağımlı değişken

# Veri setini eğitim ve test setlerine bölelim
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8, random_state=42)

# Lojistik regresyon modelini oluşturalım ve eğitelim
log_reg = LogisticRegression(max_iter=10000)
log_reg.fit(X_train, y_train)

# Eğitim seti üzerinde tahmin yapalım
train_predictions = log_reg.predict(X_train)

# Test seti üzerinde tahmin yapalım
test_predictions = log_reg.predict(X_test)

# Doğruluk skorlarını hesaplayalım
train_accuracy = accuracy_score(y_train, train_predictions)
test_accuracy = accuracy_score(y_test, test_predictions)

# Sınıflandırma raporlarını görüntüleyelim
train_classification_report = classification_report(y_train, train_predictions)
test_classification_report = classification_report(y_test, test_predictions)

# Sonuçları yazdıralım
print("Eğitim Seti Doğruluk Skoru:", train_accuracy)
print("Test Seti Doğruluk Skoru:", test_accuracy)

print("\nEğitim Seti Sınıflandırma Raporu:")
print(train_classification_report)

print("\nTest Seti Sınıflandırma Raporu:")
print(test_classification_report)
```

 * Eğitim ve test setleri için 0 sınıfı için yaklaşık %91 ve 1 sınıfı için yaklaşık %63 kesinlik değerleri var.(PRECİSİON)
 * Eğitim ve test setleri için 0 sınıfı için yaklaşık %98 ve 1 sınıfı için yaklaşık %31 duyarlılık değerleri var.

* Genel olarak, modelin 0 sınıfını tahmin etme performansı 1 sınıfını tahmin etme performansından daha yüksektir.
